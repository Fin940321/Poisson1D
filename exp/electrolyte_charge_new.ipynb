# xvg檔的作圖(學長的圖)

```python
import matplotlib.pyplot as plt
import numpy as np

# 儲存 z 座標和對應電荷密度
z = []
charge_density = []

# 讀取並解析 .xvg 檔案內容
with open("electrolyte_charge_density.xvg", "r") as f:
    for line in f:
        if not line.startswith(('#', '@')):
            parts = line.strip().split()
            if len(parts) == 2:
                z_val = float(parts[0])
                charge_val = float(parts[1])
                z.append(z_val)
                charge_density.append(charge_val)

# 畫圖
plt.figure(figsize=(10, 5))
plt.plot(z, charge_density, label='Charge Density')
plt.xlabel('Z Position (nm)')
plt.ylabel('Charge Density (e/Å³)')
plt.title('Electrolyte Charge Density Profile')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
