```python
file_path = r"/home/atuo/ion_density/2V_otf_bmim_2cnt_100ns_240411/2V_otf_bmim_2cnt_100ns_24411.log"
output_file = "charges_output.dat"  # 儲存的檔名
```

```python
def search_for_charges(file_path):
    with open(file_path) as file:
        lines = file.readlines()
    
    first_iteration_index = 0
    last_iteration_index = 0
    charges_list = []
    
    x = input("起始iteration值:")
    y = input("終止iteration值:")
    c = str(f"{x} iteration")
    d = str(f"{y} iteration")
    
    for a, line1 in enumerate(lines):
        if c in line1:
            first_iteration_index = a+1
            print(first_iteration_index)
            print(line1)
            break
    if first_iteration_index == 0:
        return "未輸入起始值"
    
    for b, line2 in enumerate(lines):
        if d in line2:
            last_iteration_index = b+1
            print(last_iteration_index)
            print(line2)
            break
    if last_iteration_index == 0:
        return "未輸入終止值"
    
    for z in range(first_iteration_index, last_iteration_index):
        if "Q_numeric , Q_analytic charges on  anode" in lines[z]:
            charge_values = lines[z].strip().split()
            if len(charge_values) > 6:  
                charges_list.append(charge_values[6])
    
    # 1.檢查指定 iteration 範圍內的陰離子電荷:           
    # if charges_list:
    #     for charges in charges_list:
    #         print(charges)
    # else:
    #     print("未找到 charges")

    # 2.把輸出的結果存入 .dat 檔案:
    if charges_list:
        with open(output_file, "w") as f:
            for charge in charges_list:
                f.write(charge + "\n")
        
        print(f"已將 {len(charges_list)} 筆 charges 儲存至 {output_file}")
    else:
        print("未找到 charges")
```

```python
search_for_charges(file_path)
```
